```python
#Import spaCy and required language modules
import spacy
```

###  Tokenization using spaCy

In this example, we use the **spaCy** library to perform *tokenization*, which is the process of breaking text into individual words or symbols (called tokens).

Steps:
1. Load a pre-trained English NLP model.
2. Define a sample sentence.
3. Process the text using spaCy.
4. Iterate over each token and print it.

This is one of the most basic steps in Natural Language Processing (NLP).

```python
nlp = spacy.load("en_core_web_sm")
text = "Apple is looking at buying a startup in the UK for $1 billion."
doc = nlp(text)

for token in doc:
    print(token.text)
```

###  Tokenization + Part-of-Speech (POS) Tagging using spaCy

In this example, we not only split the text into tokens, but also identify the **Part of Speech (POS)** of each token.

- **Token**: A word, punctuation, or symbol in the text.
- **POS Tag**: Grammatical role of the token (e.g., noun, verb, adjective).

This helps us understand the structure of a sentence.

```python
for token in doc:
    print(token.text, token.pos_)
```

###  Named Entity Recognition (NER) using spaCy

In this example, we extract **named entities** from the text.

- **Named Entity**: A real-world object such as a person, organization, location, date, or money.
- spaCy identifies these entities and assigns a **label** to each one.

Examples of entity types:
- ORG → Organization (e.g., Apple)
- GPE → Country, city (e.g., UK)
- MONEY → Monetary values (e.g., $1 billion)

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```

## Lemmatization using spaCy

In this example, we extract the **lemma** of each token.

- **Lemma**: The base or dictionary form of a word.
- It helps normalize words so that different forms are treated the same.

Examples:
- "looking" → "look"
- "buying" → "buy"
- "is" → "be"

Lemmatization is important in NLP tasks like search engines, text analysis, and machine learning.

```python
for token in doc:
    print(token.text, "->", token.lemma_)
```

###  Sentence Segmentation using spaCy

In this example, we split the text into **sentences**.

- **Sentence Segmentation**: The process of dividing a paragraph into individual sentences.
- spaCy automatically detects sentence boundaries using punctuation and language rules.

This is useful for:
- Text summarization
- Chatbots
- Document processing

```python
doc = nlp(text)
for sent in doc.sents:
    print(sent.text)
```
